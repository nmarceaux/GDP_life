```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('all_data.csv')

print(df.head())
```

```python
df = df.rename(columns={'Life expectancy at birth (years)': 'Life Expectancy'})
print(df.head())
```

```python
sns.scatterplot(data=df, x="Year", y="Life Expectancy", hue = "Country")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
plt.clf()

sns.scatterplot(data=df, x="Year", y="GDP", hue = "Country")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
plt.clf()

sns.scatterplot(data=df, x="GDP", y="Life Expectancy", hue = "Country")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
plt.clf()
```

```python
countries = df['Country'].unique()
print(countries)
```

```python
sns.boxplot(data = df, x = 'Country', y = 'GDP')
plt.xticks(rotation = 80)
plt.show()
plt.clf()

sns.boxplot(data = df, x = 'Country', y = 'Life Expectancy')
plt.xticks(rotation = 80)
plt.show()
plt.clf()
```

```python
sns.heatmap(df.corr())
```

```python
### Percent Change
## (final - initial) / intial * 100

df['gdp_pct_change'] = df['GDP'].pct_change()

print(df)
```

```python
sns.barplot(x="Year", y="gdp_pct_change", hue="Country", data=df)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.xticks(rotation = 90)
plt.show()
```
